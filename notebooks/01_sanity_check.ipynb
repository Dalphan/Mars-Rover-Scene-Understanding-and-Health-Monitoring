# Sanity Check
Load a dataset sample and save a visualization grid.

```python
from pathlib import Path
from hydra import initialize, compose
from datasets.datamodule import SegmentationDataModule
from utils.visualization import save_prediction_grid
import torch

config_dir = "../configs" if Path("../configs").exists() else "configs"
with initialize(version_base="1.3", config_path=config_dir):
    cfg = compose(config_name="config")

dm = SegmentationDataModule(cfg)
dm.setup("fit")
images, masks = next(iter(dm.train_dataloader()))
preds = torch.zeros_like(masks)
save_prediction_grid(
    images,
    masks,
    preds,
    output_dir=cfg.paths.output_dir,
    epoch=0,
    max_samples=4,
    mean=cfg.data.normalize.mean,
    std=cfg.data.normalize.std,
    ignore_index=cfg.ignore_index,
    num_classes=cfg.num_classes,
)
print("Saved a sample grid to", cfg.paths.output_dir)
```
